```python
import pandas as pd
import numpy as np

from dateutil.relativedelta import *
from dateutil.easter import *
from dateutil.rrule import *
from dateutil.parser import *
from datetime import *
```

```python
data_directory = "Data/Trip Data/"
```

```python
tripdataDF = pd.DataFrame()  

for year in np.arange(2016, 2019, 1):
    filepath1 = data_directory +"JC-"+ str(year)
    for month in np.arange(1, 13, 1):
        if(year == 2018 and month > 10):
            break
        elif(month < 10):
            filepath2 = filepath1 +"0"+ str(month) +"-citibike-tripdata.csv"
        else:
            filepath2 = filepath1 + str(month)+"-citibike-tripdata.csv"
        print(filepath2)
        df = pd.read_csv(filepath2)
        
        # Cleaning column headers
        df.columns = df.columns.str.strip().str.lower().str.replace(' ', '')
        
        # Cleaning dates and creating calculated fields for age and duration in minutes
        df['starttime'] = df['starttime'].map(lambda date : parse(date))
        df['stoptime'] = df['stoptime'].map(lambda date : parse(date))
        df['Age'] = df['birthyear'].map(lambda birthyear: (2018 - birthyear))
        df['Tripduration_minutes'] = df['tripduration'].map(lambda tripduration: round(tripduration/60,2))
        
        tripdataDF = pd.concat([tripdataDF,df])

tripdataDF.head()
```

```python
tripdataDF.reset_index(drop=True)
```

```python
tripdataDF.to_csv('Combined_TripData.csv', index_label  = 'Index')
```
